```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

# Loading data


```python
diabetes = load_diabetes()
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df
```

# Running model

```python
x_train, x_test, y_train, y_test = train_test_split(df[diabetes.feature_names], df["target"], test_size=0.5)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predicted = lr.predict(x_test)
```

```python
print(f"Mean squared error is {mean_squared_error(y_test,predicted)}")
print(f"Mean absolute error is {mean_absolute_error(y_test,predicted)}")
```

```python
plt.scatter(x = predicted, y = y_test)
```
